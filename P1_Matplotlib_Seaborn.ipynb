```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Creando un DF con los datos en csv de caudales de la estación rio pallanga del senamhi

```python
df=pd.DataFrame(pd.read_csv('rioPallangaDiario.csv'))
df.head(2)
```

Eliminar de la df las columnas que no nos sirvan

```python
# df.columns
df=df.drop(['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15',
       'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19',
       'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23',
       'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27'],axis=1)
df.columns # Para ver las categorías disponibles
#df.shape[0] # Para ver el tamaño de nuestros datos (cantidad de registros)
#np.arange(0,df.shape[0])
```

Completando los valores nan con la media

```python
# df.isnull().sum()#Averiguar la cantidad de valores nulos en el DF
df=df.fillna(df['Valor'].mean())
# df.isnull().any()
```

Creando 2 variables:
x: una serie de 0 hasta el correlativo del último valor
y: los valores de caudal

```python
x=np.arange(0,df.shape[0])
y=df['Valor']
```

Creando un DF2 con los datos en csv de caudales de la estación carhuacayan del senamhi

```python
df2=pd.DataFrame(pd.read_csv('CarhuacayanDiario.csv'))
# df2.columns
# df2.isnull().any()
```

Creando 2 variables:
x2: una serie de 0 hasta el correlativo del último valor
y2: los valores de caudal

```python
y2=df2['Valor']
x2=np.arange(0,y2.shape[0])
```

Creando un DF3 con los datos en csv de caudales de la estación Milloc del senamhi

```python
df3=pd.DataFrame(pd.read_csv('millocDiario.csv'))
# df3.columns
# df3.isnull().any()
x3=np.arange(0,df3['Valor'].shape[0])
y3=df3['Valor']
# plt.plot(x3,y3)
```

# Matplotlib

Probando los tipos de gráficas con el método Pyplot

```python
# plt.plot(x,y,'rx:')
# plt.hist(y)
# plt.scatter(x,y)
plt.boxplot(y)
plt.show()
```

Graficando con método orientado a objetos

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8, 0.6])
axes2 = fig.add_axes([0.20,0.45,0.3,0.2])
axes1.plot(x,y,'#284b96', linewidth=0.5)
axes1.set_title('Caudales registrados en E. Río Pallanga')
axes1.set_ylabel('Caudal')
axes1.set_xlabel('Día')
axes2.plot(x2,y2,'#bde0fe',linewidth=0.5)
axes2.set_title('Caudales registrados en E. Carhuacayán', fontsize = 7)
axes2.set_facecolor('#284b96')
axes2.set_xlabel('Día', fontsize = 7)
axes2.set_ylabel('caudal', fontsize = 7)
axes2.set_xticklabels(x, fontsize=7)#cambiar el tamaño de font de los valores de los ejes
axes2.set_yticklabels(y, fontsize=7)

plt.show()
```

## Barras

Comparando el N° de registros con los que cuenta cada estación usando diccionarios

```python
dictionary={'RioPallanga':df['Valor'].shape[0],'Carhuacayan':df2['Valor'].shape[0],'Milloc':df3['Valor'].shape[0]}
dictionary
```

```python
plt.bar(range(len(dictionary)),list(dictionary.values()),align='center',color='#b5179e')
plt.xticks(range(len(dictionary)),list(dictionary.keys()),fontsize=7)
plt.xlabel('Estaciones')
plt.ylabel('Cantidad de días registrados (d)')
plt.show()
```

```python
estaciones=['Río Pallanga','Carhuacayan','Milloc']
registros=[df['Valor'].shape[0],df2['Valor'].shape[0],df3['Valor'].shape[0]]
plt.bar(estaciones,registros,width=0.5,color=['#9c89b8','#f0a6ca','#efc3e6'])
plt.xticks(np.arange(3),('E.RP','E.C','E.M'),rotation=45)
plt.xlabel('Estaciones')
plt.ylabel('Cantidad de días registrados (d)')
plt.show()
```

```python
plt.barh(estaciones,registros, color=['#fec5bb','#d8e2dc','#ffe1a8'])
plt.title('Nro de registros por cada estación')
plt.xlabel('Cantidad de días registrados (d)')
plt.ylabel('Estación')
plt.show()
```
