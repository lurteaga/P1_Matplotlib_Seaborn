# Medidas de dispersión

```python
df_rioPallangaDiario['Valor'].std()
```

```python
# Rango = valor max - valor min
rango = df_rioPallangaDiario['Valor'].max() - df_rioPallangaDiario['Valor'].min()
rango
```

```python
# Quartiles
median = df_rioPallangaDiario['Valor'].median()
Q1 = df_rioPallangaDiario['Valor'].quantile(q=0.25)
Q3 = df_rioPallangaDiario['Valor'].quantile(q=0.75)
min_val = df_rioPallangaDiario['Valor'].quantile(q=0)
max_val = df_rioPallangaDiario['Valor'].quantile(q=1.0)
print(min_val, Q1, median, Q3, max_val)
```

```python
iqr = Q3 - Q1
iqr
```

```python
#Límites para detección de outliers (en datos simétricamente distribuidos)
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print('rango para detección de outliers: {}, {}'.format(minlimit, maxlimit))
#Vemos que el valor mínimo es negativo y esto es porque no son datos simétricamente distribuidos
```

```python
sns.set(rc={'figure.figsize':(11,5)})
f, (ax_hist, ax_box) = plt.subplots(nrows=1,ncols=2, sharex=False, gridspec_kw={"width_ratios": (.6, .4)})
sns.histplot(df_rioPallangaDiario['Valor'], ax=ax_hist)
sns.boxplot(df_rioPallangaDiario['Valor'], ax=ax_box)
ax_hist.set(xlabel='');
```
